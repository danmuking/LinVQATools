# 相似度分析

```python
import os
```

```python
root = '/data/ly/unique/VQA_ODV'
video_list = [os.path.join(root,i) for i in os.listdir(root)]
temp=[]
for i in video_list:
    for j in os.listdir(i):
        temp.append(os.path.join(i,j))
video_list = temp
```

```python
clip_num = []
for video_dir in video_list:
    clip_num.append((video_dir,len(os.listdir(video_dir))))
```

```python
clip_num = sorted(clip_num,key=lambda i: i[1])
print(clip_num)
```
